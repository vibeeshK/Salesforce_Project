Calculating driving costs and time

```python
import pandas as pd
```

```python
state_gas = pd.read_csv("gas price by state.csv")
road = pd.read_csv('road_distance.csv')
state_gas
```

```python
# Merge employee data with gas prices
road_dist_with_price = pd.merge(road, state_gas, left_on="state", right_on="state")
road_dist_with_price
```

```python
# Merge gas prices with the prev dataframe

road_gas_prices = road_dist_with_price[['Atlanta',	'Bellevue',	'Chicago',	'Dallas',	'Denver',	'Indianapolis',	'New York',	'San Francisco',	'Seattle',	'The Ranch']].divide(24.2).multiply(road_dist_with_price['Regular'], axis="index")
road_gas_prices['postal_code'] = road_dist_with_price['postal_code']
road_gas_prices['state'] = road_dist_with_price['state']
road_gas_prices['latitude'] = road_dist_with_price['latitude']
road_gas_prices['longitude'] = road_dist_with_price['longitude']
```

```python
road_gas_prices.to_csv('road_cost.csv')
```

```python
# Calculating driving time assuming 32 mph:
road_time = road_gas_prices
road_time['Atlanta'] = road_time['Atlanta']/32
road_time['Bellevue'] = road_time['Bellevue']/32
road_time['Chicago'] = road_time['Chicago']/32
road_time['Dallas'] = road_time['Dallas']/32
road_time['Denver'] = road_time['Denver']/32
road_time['Indianapolis'] = road_time['Indianapolis']/32
road_time['New York'] = road_time['New York']/32
road_time['San Francisco'] = road_time['San Francisco']/32
road_time['Seattle'] = road_time['Seattle']/32
road_time['The Ranch'] = road_time['The Ranch']/32
```

```python
road_time.to_csv('road_time.csv')
```
